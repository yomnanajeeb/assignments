```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def tanh(x):
    return math.tanh(x)

Wf, Whf, bf = 0.5, 0.1, 0
Wi, Whi, bi = 0.6, 0.2, 0
Wc, Whc, bc = 0.7, 0.3, 0
Wo, Who, bo = 0.8, 0.4, 0

# Output layer
Wy, by = 4, 0

# Input sequence
X = [1, 2, 3]

# Initial states
h = 0
c = 0

for t, x in enumerate(X, start=1):
    print(f"Time Step t = {t}, Input x{t} = {x}")

    # Forget gate
    f = sigmoid(Wf * x + Whf * h + bf)
    print(f"Forget gate f{t} = {f:.3f}")

    # Input gate
    i = sigmoid(Wi * x + Whi * h + bi)
    print(f"Input gate i{t} = {i:.3f}")

    # Candidate cell state
    c_tilde = tanh(Wc * x + Whc * h + bc)
    print(f"Candidate cell state c_tilde{t} = {c_tilde:.3f}")

    # Cell state update
    c = f * c + i * c_tilde
    print(f"Cell state c{t} = {c:.3f}")

    # Output gate
    o = sigmoid(Wo * x + Who * h + bo)
    print(f"Output gate o{t} = {o:.3f}")

    # Hidden state update
    h = o * tanh(c)
    print(f"Hidden state h{t} = {h:.3f}\n")

x4 = 4
print("Time Step t = 4, Predict x4\n")

f4 = sigmoid(Wf * x4 + Whf * h + bf)
i4 = sigmoid(Wi * x4 + Whi * h + bi)
c_tilde4 = tanh(Wc * x4 + Whc * h + bc)
c4 = f4 * c + i4 * c_tilde4
o4 = sigmoid(Wo * x4 + Who * h + bo)
h4 = o4 * tanh(c4)

print(f"Forget gate f4 = {f4:.3f}")
print(f"Input gate i4 = {i4:.3f}")
print(f"Candidate cell state c_tilde4 = {c_tilde4:.3f}")
print(f"Cell state c4 = {c4:.3f}")
print(f"Output gate o4 = {o4:.3f}")
print(f"Hidden state h4 = {h4:.3f}")

# Final prediction
y_hat = Wy * h4 + by
print(f"\nPredicted value y_hat = {y_hat:.3f}")
```
